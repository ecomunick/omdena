## Inferential Statistics (6 marks):
Formulate a hypothesis related to the dataset.
Conduct a statistical test (e.g., t-test) to test the hypothesis.
Provide conclusions based on the test results.

- Null Hypothesis (H0): There is no significant difference in the happiness scores between countries with high levels of GDP per capita and countries with low levels of GDP per capita.

- Alternative Hypothesis (H1): There is a significant difference in the happiness scores between countries with high levels of GDP per capita and countries with low levels of GDP per capita.

To test this hypothesis, I splitted the dataset into two groups based on the `GDP per capita`, then performed a `t-test` to compare the mean `happiness scores` between these two groups.

````python
from scipy.stats import ttest_ind

# Split the dataset into two groups based on GDP per capita (e.g., using median as the threshold)
median_gdp_per_capita = df['GDP per capita'].median()
high_gdp_countries = df[df['GDP per capita'] >= median_gdp_per_capita]
low_gdp_countries = df[df['GDP per capita'] < median_gdp_per_capita]

# Perform a t-test to compare the mean happiness scores between the two groups
t_statistic, p_value = ttest_ind(high_gdp_countries['Happiness score'], low_gdp_countries['Happiness score'])

# Set the significance level (alpha)
alpha = 0.05

# Print the test results
print("T-test Results:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Check if the p-value is less than alpha to determine significance
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in the happiness scores between countries with high and low levels of GDP per capita.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in the happiness scores between countries with high and low levels of GDP per capita.")
````

## Conclusion:
The p-value (7.18e-21) is less than the chosen significance level (alpha = 0.05), we reject the null hypothesis and consider that there is a significant difference in the happiness scores between countries with high and low levels of GDP per capita.

###Outlier Detection (5 marks):
- Identify and handle outliers in the dataset.
- Explain the methodology and reasoning behind your approach.


````python
# Importing libraries
import numpy as np

# Function to detect and handle outliers using IQR method
def detect_handle_outliers_iqr(data, threshold=1.5):
    outliers_indices = []
    for column in data.columns:
        if data[column].dtype in ['int64', 'float64']:
            Q1 = np.percentile(data[column], 25)
            Q3 = np.percentile(data[column], 75)
            IQR = Q3 - Q1
            lower_bound = Q1 - threshold * IQR
            upper_bound = Q3 + threshold * IQR
            outliers_indices.extend(data[(data[column] < lower_bound) | (data[column] > upper_bound)].index)
    outliers_indices = list(set(outliers_indices))
    return outliers_indices

# Detect and handle outliers in the dataset
outliers_indices = detect_handle_outliers_iqr(df)

# Remove outliers from the dataset
df_cleaned = df.drop(outliers_indices)

# Explanation:
# - In this approach, we calculate the Interquartile Range (IQR) for each numerical variable in the dataset.
# - Outliers are identified as values lying outside the range of Q1 - 1.5 * IQR to Q3 + 1.5 * IQR.
# - Detected outliers are either removed or replaced with a central value (e.g., mean or median).
# - The threshold parameter in the detect_handle_outliers_iqr function can be adjusted based on the dataset characteristics and domain knowledge.
# - This method is robust to outliers and provides a systematic way to handle them.

# Display the cleaned dataset
print("Number of outliers removed:", len(outliers_indices))
print("Shape of cleaned dataset:", df_cleaned.shape)
````
